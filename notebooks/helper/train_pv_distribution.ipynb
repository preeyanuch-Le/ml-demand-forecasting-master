```python
end_date = date.today().strftime("%Y-%m-%d")
one_year = date.today() - relativedelta(months=+12)
start_date = one_year.strftime("%Y-%m-%d")
```

```python
hal = Hal()

sql = f"""
SELECT fpp.id_product,
        CASE WHEN fpp.id_shop = 5 THEN 'ID' 
              WHEN fpp.id_shop = 2 THEN 'MY'
              WHEN fpp.id_shop = 11 THEN 'MY'
              WHEN fpp.id_shop = 14 THEN 'MY'
              ELSE 'TH' END as warehouse
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 0 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 6 THEN fpp.counter ELSE 0 END) as week1
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 7 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 13 THEN fpp.counter ELSE 0 END) as week2
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 14 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 20 THEN fpp.counter ELSE 0 END) as week3
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 21 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 27 THEN fpp.counter ELSE 0 END) as week4
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 28 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 34 THEN fpp.counter ELSE 0 END) as week5
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 35 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 41 THEN fpp.counter ELSE 0 END) as week6
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 42 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 48 THEN fpp.counter ELSE 0 END) as week7
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 49 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 55 THEN fpp.counter ELSE 0 END) as week8
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 56 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 62 THEN fpp.counter ELSE 0 END) as week9

,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 63 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 69 THEN fpp.counter ELSE 0 END) as week10
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 70 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 76 THEN fpp.counter ELSE 0 END) as week11
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 77 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 83 THEN fpp.counter ELSE 0 END) as week12
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 84 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 90 THEN fpp.counter ELSE 0 END) as week13
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 91 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 97 THEN fpp.counter ELSE 0 END) as week14
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 98 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 104 THEN fpp.counter ELSE 0 END) as week15
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 105 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 111 THEN fpp.counter ELSE 0 END) as week16
,SUM(CASE WHEN DATEDIFF(days,dp.date_released ,fpp.date_view) >= 112 AND DATEDIFF(days,dp.date_released ,fpp.date_view) <= 118 THEN fpp.counter ELSE 0 END) as week17
 
FROM dwh.fact_product_pageview fpp 
LEFT JOIN dwh.dim_product dp 
ON dp.id_product = fpp.id_product

WHERE dp.date_released BETWEEN '{start_date}' AND '{end_date}'
AND dp.parent_product_line NOT IN ('Free Gift')
AND dp.product_line NOT IN('Bags','3P Apparel','Shoes')
AND dp.henry_category_1 NOT IN ('Accessories','Bags','Bath&Body','Beverage Container','Cosmetics','Hair','Miscellaneous','Shoes','Skin Care','Stationery')
AND dp.product_cost_usd_first_order_date > 0 
AND dp.brand IN ('Alita','Basics','Pomelo', 'BEET by Pomelo', 'Holiday Collection', 'Pomelo x Tex Saverio','Blackdog BKK')
AND dp.release_collection_name IS NOT NULL
Group by 1,2

"""
```

```python
pv_dist = hal.get_pandas_df(sql)
```

```python
pv_dist
```

```python
pv_dist = pv_dist.groupby("warehouse").sum().reset_index()
del pv_dist["id_product"]
pv_dist
```

```python
pv_dist2 = pv_dist.iloc[:, :]
```

```python
pv_dist_unpivot = pv_dist2.melt(
    id_vars=["warehouse"], var_name="week_id", value_name="pv"
)
pv_dist_unpivot
```

```python
# find totoal pageview
pv_tot = (
    pv_dist_unpivot.groupby("warehouse")["pv"]
    .sum()
    .reset_index()
    .rename(columns={"pv": "tot_pv"})
)
```

```python
df = pv_dist_unpivot.merge(pv_tot[["warehouse", "tot_pv"]], on="warehouse")
df
```

```python
df["pv_dist"] = df["pv"] / df["tot_pv"]
```
