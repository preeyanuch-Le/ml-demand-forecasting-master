```python
dictionary = {
    "week10": "weeks10",
    "week11": "weeks11",
    "week12": "weeks12",
    "week13": "weeks13",
    "week14": "weeks14",
    "week15": "weeks15",
    "week16": "weeks16",
    "week17": "weeks17",
}
for key in dictionary.keys():
    df_pv_final["week_id"] = df_pv_final["week_id"].replace(key, dictionary[key])

df_pv_final = df_pv_final.sort_values(by="week_id")
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=3,
    cols=1,
    shared_xaxes=True,
    vertical_spacing=0.02,
    subplot_titles=("TH", "MY", "ID"),
)

fig.add_trace(
    go.Scatter(
        x=df_pv_final[df_pv_final["warehouse"] == "ID"]["week_id"],
        y=df_pv_final[df_pv_final["warehouse"] == "ID"]["lookbook_pv_dist"],
    ),
    row=3,
    col=1,
)

fig.add_trace(
    go.Scatter(
        x=df_pv_final[df_pv_final["warehouse"] == "MY"]["week_id"],
        y=df_pv_final[df_pv_final["warehouse"] == "MY"]["lookbook_pv_dist"],
    ),
    row=2,
    col=1,
)

fig.add_trace(
    go.Scatter(
        x=df_pv_final[df_pv_final["warehouse"] == "TH"]["week_id"],
        y=df_pv_final[df_pv_final["warehouse"] == "TH"]["lookbook_pv_dist"],
    ),
    row=1,
    col=1,
)

fig.update_layout(height=900, width=900, title_text="lookbook_pv_dist by warehouse")
fig.update_xaxes(type="category")
fig.update_xaxes(categoryorder="category ascending")

fig.write_html(
    f"/home/ec2-user/SageMaker/business-intelligence-notebooks/dfm_clothing/online_dfm_v2/plot/plot_lookbook_pv_dist_deploy.html"
)
```
